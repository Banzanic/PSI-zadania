# Wielowymiarowy rozkład normalny

```python
import matplotlib.pylab as plt
import numpy as np
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.mlab as mlab
from matplotlib import cm
from scipy.stats import multivariate_normal
```

# Zad. 
Narysujmy gęstość dwuwymiarowego rozkładu Normalnego o parametach 

* n=[0,0], cov = [[1,0],[0,1]]
* n=[0,0], cov = [[2,0],[0,1]]
* n=[0,0], cov = [[1,0],[0,2]]

```python
x=np.linspace(-5,5,100)
y=np.linspace(-5,5,100)
X, Y=np.meshgrid(x,y)
pos=np.empty(X.shape + (2,))
pos[:, :, 0]=X; pos[:, :, 1] = Y
rv1=multivariate_normal([0, 0], [[1, 0], [0, 1]])
rv2=multivariate_normal([0, 0], [[2, 0], [0, 1]])
rv3=multivariate_normal([0, 0], [[1, 0], [0, 2]])
fig=plt.figure()
ax=fig.gca(projection='3d')
ax.plot_surface(X, Y, rv1.pdf(pos),cmap='viridis',linewidth=0)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

```python
fig=plt.figure()
ax=fig.gca(projection='3d')
ax.plot_surface(X, Y, rv2.pdf(pos),cmap='viridis',linewidth=0)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

```python
fig=plt.figure()
ax=fig.gca(projection='3d')
ax.plot_surface(X, Y, rv3.pdf(pos),cmap='viridis',linewidth=0)
ax.set_xlabel('x')
ax.set_ylabel('y')
ax.set_zlabel('z')
plt.show()
```

# Zad 
Wylosujmy próbkę z dwuwymiarowego rozkładu Normalnego o parametach 

* n=[0,0], cov = [[1,0],[0,1]]
* n=[0,0], cov = [[2,0],[0,1]]
* n=[0,0], cov = [[1,0],[0,2]]

i narysuj odpowiedni wykres

```python
n=[0, 0]
cov1=[[1, 0], [0, 1]]
cov2=[[2, 0], [0, 1]]
cov3=[[1, 0], [0, 2]]
```

```python
data=np.random.multivariate_normal(n, cov1, 5000)
plt.plot(data[:, 0], data[:, 1], '.')
```

```python
data=np.random.multivariate_normal(n, cov2, 5000)
plt.plot(data[:, 0], data[:, 1], '.')
```

```python
data=np.random.multivariate_normal(n, cov3, 5000)
plt.plot(data[:, 0], data[:, 1], '.')
```
